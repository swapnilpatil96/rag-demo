```python
text = """
Company Leave Policy:
Employees are entitled to 26 paid leaves per year.
Out of 26, privilege leaves are 18 and casual leaves are 8.
Also Maternity and Paternity leaves are 10 and 5 respectively.
Only maximum 45 leaves can be carry forwarded to next year.
Working hours are from 9 AM to 6 PM.
Casual leave requires manager approval.
"""

with open("leave-policy-details.txt", "w") as f:
    f.write(text)
```

```python
!pip install -U langchain langchain-community
```

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("leave-policy-details.txt")
documents = loader.load()

documents
```

```python
!pip install -U langchain-text-splitters
```

```python
from langchain_text_splitters import CharacterTextSplitter

splitter = CharacterTextSplitter(chunk_size=200, chunk_overlap=20)
docs = splitter.split_documents(documents)

docs
```

```python
from langchain_community.embeddings import OpenAIEmbeddings

embeddings = OpenAIEmbeddings()
```

```python
from langchain_community.vectorstores import FAISS

vectorstore = FAISS.from_documents(docs, embeddings)
```

```python
query = "how many leaves can be carry forwarded?"

retrieved_docs = vectorstore.similarity_search(query, k=2)

retrieved_docs
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI()

context = "\n".join([doc.page_content for doc in retrieved_docs])

prompt = f"""
Answer the question based only on the context below:

Context:
{context}

Question:
{query}
"""

response = llm.invoke(prompt)

print(response.content)
```
